```python
X = d.drop(columns=['target'])
y = d['target']
```

```python
from sklearn.preprocessing import StandardScaler

numerical_cols = ['x1', 'x2', 'x3', 'x4', 'x5', 'x6', 'x7']
scaler = StandardScaler()
X_train[numerical_cols] = scaler.fit_transform(X_train[numerical_cols])
X_test[numerical_cols] = scaler.transform(X_test[numerical_cols])

print(X_train.head())
print(X_test.head())
```

```python
from sklearn.ensemble import RandomForestRegressor

rf_model = RandomForestRegressor(random_state=42)
rf_model.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

y_pred = rf_model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
mae_rf = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Random Forest Regression: ")
print()
print("MSE: ", mse)
print("MAE: ",mae_rf)
print("R2 Score: ", r2)
```

```python
from sklearn.linear_model import LinearRegression

linear_reg_model = LinearRegression()
linear_reg_model.fit(X_train, y_train)
```

```python
linear_reg_y_pred = linear_reg_model.predict(X_test)
mse_lr = mean_squared_error(y_test, linear_reg_y_pred)
mae = mean_absolute_error(y_test, linear_reg_y_pred)
r2_lr = r2_score(y_test, linear_reg_y_pred)
print("Linear Regression: ")
print()
print("MSE lr: ", mse_lr)
print("MAE: ",mae)
print("R2 Score lr: ", r2_lr)
```

```python
from sklearn.tree import DecisionTreeRegressor

dt_model = DecisionTreeRegressor(random_state=42)
dt_model.fit(X_train, y_train)
```

```python
dt_y_pred = dt_model.predict(X_test)
mse_dt = mean_squared_error(y_test, dt_y_pred)
mae_dt = mean_absolute_error(y_test, dt_y_pred)
r2_dt = r2_score(y_test, dt_y_pred)
print("Decision Tree: ")
print()
print("MSE DT: ", mse_dt)
print("MAE: ", mae_dt)
print("R2 Score DT: ", r2_dt)
```

```python
from xgboost import XGBRegressor

xgb_model = XGBRegressor(random_state=42)
xgb_model.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

xgb_y_pred = xgb_model.predict(X_test)
mse_xgb = mean_squared_error(y_test, xgb_y_pred)
mae_xg = mean_absolute_error(y_test, xgb_y_pred)
r2_xgb = r2_score(y_test, xgb_y_pred)

print("XGBoost Regression:")
print()
print("MSE: ",mse_xgb)
print("MAE: ", mae_xg)
print("R2 Score: ",r2_xgb)
```
